```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter.scrolledtext import *
import tkinter.filedialog
import os 
import string
from Crypto.Cipher import AES
from datetime import datetime
```

```python
window = Tk()
window.title("AES Encryption GUI")
window.geometry("700x400")

def encrypt():
    AES_KEY = os.urandom(32)
    AES_IV = os.urandom(16)

    with open(x, 'rb') as fd:
        contents = fd.read()
    filename = f"aes_key_iv_{datetime.now().isoformat()[:-7].replace(':','')}.txt"
    with open(filename, "wb") as f:
        f.write(AES_KEY + b"\n")
        f.write(AES_IV + b"\n")
        
    encryptor = AES.new(AES_KEY, AES.MODE_CFB, AES_IV)
    encrypted = encryptor.encrypt(contents)
    encrypted_file = f"encrypted_data_{datetime.now().isoformat()[:-7].replace(':','')}.txt"
    with open(encrypted_file, "wb") as f:
        f.write(encrypted)

def openfiles():
    filename = filedialog.askopenfilename(initialdir = "/",title = "Select a File",filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
    global x
    x = filename
    label0.configure(text=filename)
    print(x)
    
def openfiles1():
    filename = filedialog.askopenfilename(initialdir = "/",title = "Select a File",filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
    global y
    y = filename
    label3.configure(text=y)
    
def keyopen():
    aes_key_iv_file = filedialog.askopenfilename(initialdir = "/",title = "Select a File",filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
    with open(aes_key_iv_file, "rb") as f:
        global KEY
        global IV
        KEY = f.readline().strip()
        IV = f.readline().strip()
    label2.configure(text=aes_key_iv_file)
    
#encryption part
label0 = Label(window,text = "Upload file",width = 50, height = 4,fg = "blue")
label0.grid(column=0, row=0)
b0=Button(window,text="select the file for encryption", width=20,command=openfiles,bg='#c5cae9')
b0.grid(row=3,column=0,padx=10,pady=10)
b1 = Button(window,text="encrypt",width=12,command=encrypt,bg='#c5cae9')
b1.grid(row=3,column=1,padx=10,pady=10)

#upload key
label2 =  Label(window,text = "key",width = 50, height = 4,fg = "blue")
label2.grid(column=0, row=4)
b2=Button(window,text="select the key file", width=20,command=keyopen,bg='#c5cae9')
b2.grid(row=6,column=0,padx=10,pady=10)


label3 =  Label(window,text = "Upload encryption file",
                            width = 100, height = 4,
                            fg = "blue")
label3.grid(column=0, row=7)


b4=Button(window,text="select the file for decryption", width=20,command=openfiles1,bg='#c5cae9')
b4.grid(row=8,column=0,padx=10,pady=10)
# key_file = Text(window,height=1)# Initial was Text(tab2)
# key_file.grid(row=7,column=1, columnspan=3,padx=5,pady=3)

# label3 = Label(window,text='enter aes iv',padx=5,pady=5)
# label3.grid(column=0,row=8)

# key_file1 = Text(window,height=1)# Initial was Text(tab2)
# key_file1.grid(row=8,column=1, columnspan=3,padx=5,pady=3)
def decrypt():
    AES_KEY = KEY
    AES_IV = IV
    with open(y, 'rb') as fd:
        contents = fd.read()
    decryptor = AES.new(AES_KEY, AES.MODE_CFB,AES_IV)
    decrypted = decryptor.decrypt(contents)
    print(decryptor)
    with open('decrypted.txt', 'wb') as fd:
        fd.write(decrypted)


b4 = Button(window,text="decrypt",width=12,command=decrypt,bg='#c5cae9')
b4.grid(row=9,column=0,padx=10,pady=10)

window.mainloop()
```
