```python
import tkinter as tk
from tkinter import *
from tkinter import ttk
from tkinter.scrolledtext import *
import tkinter.filedialog
import os 
import string
from Crypto.Cipher import AES
```

```python
window = Tk()
window.title("AES Encryption GUI")
window.geometry("700x400")
label1 = Label(window,
                            text = "Upload file",
                            width = 50, height = 4,
                            fg = "blue")
label1.grid(column=0, row=0)

def openfiles():
    filename = filedialog.askopenfilename(initialdir = "/",title = "Select a File",filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
    global x
    x = filename
    label1.configure(text=filename)
    print(x)
    
    
def openfiles1():
    filename = filedialog.askopenfilename(initialdir = "/",title = "Select a File",filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
    label2.configure(text="File Opened: "+filename)

b0=Button(window,text="select the file for encryption", width=20,command=openfiles,bg='#c5cae9')
b0.grid(row=3,column=0,padx=10,pady=10)


def encrypt():
    with open(x, 'rb') as fd:
        contents = fd.read()
    AES_KEY = os.urandom(32)#32 bytes aes key
    AES_IV = os.urandom(16)
    with open('aes_key.txt', 'wb') as fd:
        fd.write(AES_KEY)
    with open('aes_iv.txt','wb') as fd:
        fd.write(AES_IV)
    encryptor = AES.new(AES_KEY, AES.MODE_CFB, AES_IV)
    encrypted = encryptor.encrypt(contents)
    with open('encrypted_txt.txt', 'wb') as fd:
        fd.write(encrypted)

b1 = Button(window,text="encrypt",width=12,command=encrypt,bg='#c5cae9')
b1.grid(row=3,column=1,padx=10,pady=10)


label2 =  Label(window,
                            text = "Upload file",
                            width = 50, height = 4,
                            fg = "blue")
label2.grid(column=0, row=4)
b2=Button(window,text="select the file for decryption", width=20,command=openfiles1,bg='#c5cae9')
b2.grid(row=6,column=0,padx=10,pady=10)
label3 = Label(window,text='enter aes key',padx=5,pady=5)
label3.grid(column=0,row=7)
key_file = Text(window,height=1)# Initial was Text(tab2)
key_file.grid(row=7,column=1, columnspan=3,padx=5,pady=3)
label3 = Label(window,text='enter aes iv',padx=5,pady=5)
label3.grid(column=0,row=8)
key_file1 = Text(window,height=1)# Initial was Text(tab2)
key_file1.grid(row=8,column=1, columnspan=3,padx=5,pady=3)

def decrypt():
    AES_KEY = key_file.get("1.0","end-1c")
    AES_IV = key_file1.get("1.0","end-1c")
    with open('encrypted_txt.txt', 'rb') as fd:
        contents = fd.read()
    iv = AES_IV[:16]
    decryptor = AES.new(AES_KEY, AES.MODE_CFB,iv)
    decrypted = decryptor.decrypt(contents)
    print(decryptor)
    with open('decrypted.txt', 'wb') as fd:
        fd.write(decrypted)
    
b3 = Button(window,text="decrypt",width=12,command=decrypt,bg='#c5cae9')
b3.grid(row=9,column=1,padx=10,pady=10)




window.mainloop()
```
